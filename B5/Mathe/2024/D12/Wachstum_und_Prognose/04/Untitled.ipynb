```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.rcParams["figure.figsize"] = (10,8)
```

```python
df = pd.DataFrame({'Zeit':[0,1.5,3,4.5, 6, 7.5, 9],
                  'Giftstoffmenge':[200, 144, 104, 75, 53, 39, 28]})
```

```python
plt.plot(df.Zeit, df.Giftstoffmenge,'x')
plt.xlabel('Zeit (Std.)')
plt.ylabel('Giftstoff Menge (mg)')
xx= np.linspace(0,9,100)
plt.plot(xx, 200* 0.72**(xx/1.5))


plt.hlines(100,0,3.165, ls='--', color='k')
plt.vlines(3.165, 100, 25, ls='--', color='k')


plt.hlines(50,0,2*3.165, ls='--', color='k')
plt.vlines(2*3.165, 50, 25, ls='--', color='k')

plt.xticks(np.arange(0,10,1)) 

plt.savefig('p3.pdf', dpi=300, bbox_inches='tight')
```
